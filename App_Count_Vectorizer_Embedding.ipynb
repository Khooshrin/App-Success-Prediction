```python
import numpy
import pandas as pd
from pandas import DataFrame
```

```python
data = pd.read_csv("books.csv")
```

```python
desc=data['description'].tolist()
```

```python
print(len(desc))
```

```python
from sklearn.feature_extraction.text import CountVectorizer
```

```python
vectorizer = CountVectorizer()
```

```python
vectorizer.fit(desc)
```

```python
print("Vocabulary: ", vectorizer.vocabulary_)
```

```python
desc_embedding = vectorizer.transform(desc)
```

```python
print(desc_embedding)
```

```python
print(desc_embedding.toarray())
```

```python
sum=data['summary'].tolist()
name=data['name'].tolist()
```

```python
vectorizer.fit(sum)
sum_embedding = vectorizer.transform(sum)
print(sum_embedding)
print(sum_embedding.toarray())
```

```python
vectorizer.fit(name)
name_embedding = vectorizer.transform(name)
print(name_embedding)
print(name_embedding.toarray())
```
